## Data Modelling

```python
import pandas as pd
```

```python
reaction = pd.read_csv('./Reactions.csv')
reaction
```

```python
reaction.drop(columns=['Unnamed: 0','Unnamed: 4','Unnamed: 5'],inplace=True)
```

```python
reaction.nunique()
```

```python
content = pd.read_csv('./Content.csv')
content
```

```python
content.drop(columns=['Unnamed: 0','User ID'],inplace=True)
```

```python
reaction.nunique()
```

```python
reactiontype = pd.read_csv('./ReactionTypes.csv')
reactiontype
```

```python
reactiontype.drop(columns='Unnamed: 0',inplace=True)
```

```python
merged_data = pd.merge(reaction, content, on='Content ID')
merged_data
```

```python
merged_data.rename(columns={'Type_x':'Type'},inplace=True)
```

```python
merged_data = pd.merge(merged_data, reactiontype, on='Type')
```

```python
merged_data
```

```python
merged_data.tail()
```

```python

# Group by 'Category' and calculate the sum of scores
category_totals = merged_data.groupby('Category')['Score'].sum().reset_index()
```

```python
category_totals
```

```python
category_totals = category_totals.sort_values(by='Score', ascending=False)
```

```python
category_totals
```

```python
top_5_categories = category_totals.head(5)
top_5_categories
```
